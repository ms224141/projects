```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Ecommerce Customers.csv")
```

```python
df.head(100)
```

```python
# as we know there are multiple features that are not directly affecting the feature we have to predict so we can drop these features
# as we know there are multiple features that are not directly affecting the feature we have to predict so we can drop these features
df = df.drop(["Email", "Address", "Avatar"], axis=1) # Specify axis=1 for columns and reassign to df
```

```python
# now visualize data again
df.head()
```

```python
#  now we will check if thre are any missing values in the data set ir not
df.isnull().sum()
#  no null values is present in the data set  so we dont have to do any task on that
```

```python
#  now check for correlation of the featurs
df.corr()
```

```python
x=df[["Avg. Session Length","Time on App","Length of Membership"]]
x
```

```python
y=df["Yearly Amount Spent"]
y
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Correct the variable assignment order
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)

# Now, the shapes of x_train and y_train should be consistent,
# and the model fitting should work without errors.
```

```python
model=KNeighborsRegressor(n_neighbors=5,algorithm="ball_tree")
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
#  now we can do hyperparameter tuning
#  now we can do hyperparameter tuning
from sklearn.model_selection import GridSearchCV
params={
    "n_neighbors":[3,5,6,4,8,19,10],
    "algorithm":["ball_tree","kd_tree"]
}
grid_model=GridSearchCV(estimator=model,param_grid=params)


grid_model=GridSearchCV(estimator=model,param_grid=params,cv=5)
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
y_grid=grid_model.best_estimator_.predict(x_test)
```

```python
r2_score(y_test,y_grid)
```
