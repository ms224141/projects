```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Housing.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define the order for the 'mainroad' feature
order = ['yes', 'no']

# Create the OrdinalEncoder with the correct categories
encoding = OrdinalEncoder(categories=[order])  # Use [order] instead of [order]*6

# Fit and transform the 'mainroad' column
mainroad_encoded = encoding.fit_transform(df[['mainroad']])

# Optionally, update the 'mainroad' column in the DataFrame with the encoded values
df['mainroad_encoded'] = mainroad_encoded
encoding=OrdinalEncoder(categories=[order])
df["guestroom_encoded"]=encoding.fit_transform(df[["guestroom"]])

encoding = OrdinalEncoder(categories=[order])
df["basement_encoded"] = encoding.fit_transform(df[["basement"]])
encoding=OrdinalEncoder(categories=[order])
df["hotwaterheating_encoded"]=encoding.fit_transform(df[["hotwaterheating"]])
encoding=OrdinalEncoder(categories=[order])
df["airconditioning_encoded"]=encoding.fit_transform(df[["airconditioning"]])

```

```python
df
```

```python
df.drop(["mainroad","guestroom","basement","hotwaterheating","airconditioning"],axis=1)
```

```python
#  now check for null values
df.isnull().sum()
```

```python
x = df[["area","bedrooms","bathrooms","stories","mainroad_encoded","guestroom_encoded","basement_encoded","hotwaterheating_encoded","airconditioning_encoded"]]
```

```python
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor()
model2.fit(X_train,y_train)
```

```python
decision_pred=model2.predict(X_test)
r2_score(y_test,decision_pred)
```

```python
from sklearn.svm import SVC
model3=SVC()
model3.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10,11,12,13,5,4,3,2,82], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(model3, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```
